# Energy Consumption

Data Source: https://ourworldindata.org/grapher/primary-sub-energy-source

```python
df1 = pd.read_csv("primary-sub-energy-source.csv")
df1 = df1.rename(columns={"Entity":"Country","Wind Consumption - TWh":"Wind","Hydro Consumption - TWh":"Hydro", 
                      "Solar Consumption - TWh":"Solar", 'Nuclear Consumption - TWh': "Nuclear",
                        "Biofuels Consumption - TWh - Total": "Biofuels" , 'Geo Biomass Other - TWh': 'Geo Biomass Other',
                       'Coal Consumption - TWh': 'Coal', 'Oil Consumption - TWh': 'Oil','Gas Consumption - TWh':'Gas'})
df1
```

```python
countries_list = ["United States", "France", "Germany", "China", "Japan"]
year_list=[2020]
df1.filt = df1.loc[df1['Country'].isin(countries_list) & df1["Year"].isin(year_list)]
df1.filt

df1.filt['Biofuels'].fillna(0,inplace = True)
df1.filt
```

```python
df1.filt["Fossil fuel"] = df1.filt['Coal'] + df1.filt['Oil'] + df1.filt['Gas']
df1.filt["Nuclear"] = df1.filt["Nuclear"]
df1.filt['Renewable'] = df1.filt['Wind'] + df1.filt['Hydro'] +df1.filt['Solar'] +df1.filt["Biofuels"] +df1.filt['Geo Biomass Other']
df1.filt
```

```python
df1.filt_1 = df1.filt.drop(['Code','Year', 'Wind','Hydro','Solar','Biofuels','Geo Biomass Other','Coal','Oil', 'Gas'],axis=1)
df1.filt_1=df1.filt_1.set_index ('Country')
df1.filt_1

df1.filt_100=df1.filt_1.div(df1.filt_1.sum(axis=1), axis=0)*100
df1.filt_100
```

```python
df1.filt_100 = df1.filt_100[['Fossil fuel', 'Nuclear', 'Renewable']]
df1.filt_100
```

```python
colors = ['red','green','blue']
df1.filt_100.plot(kind='barh', y=['Fossil fuel','Nuclear','Renewable'],figsize=(10,6), stacked=True, color=colors, alpha=0.7)

plt.title("Energy Consumption by Source in 2020",fontweight="bold")
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.ylabel('Country')


for n, x in enumerate([*df1.filt_100.index.values]):
    for (proportion, count, y_loc) in zip(df1.filt_100.loc[x],
                                          df1.filt_100.loc[x],
                                          df1.filt_100.loc[x].cumsum()):
                
        plt.text(x=(y_loc - proportion) + (proportion/2),
                 y=n - 0.05,
                 s=f'{np.round(proportion , 1)}%', 
                 color="black",
                 fontsize=12)
```

We can observe that for United States, Only 10% of our total energy comes from renewables and about 80% of our energy comes from fossil fuels. Comparing it to France, only half of their energy consumption comes from fossil fuel.

#### **Public Opinion of Nuclear Energy in 2011**

```python
df_opinion2011=pd.read_csv("public-opposition-to-nuclear-energy-production.csv")
df_opinion2011
```

```python
df_opinion2011 = df_opinion2011.rename(columns={"Entity":"Country", 
                                                "Opposed to nuclear energy (Ipsos MORI (2011))": "Opposed to Nuclear Energy" })
df_opinion2011['Support to Nuclear Energy']= 100- df_opinion2011['Opposed to Nuclear Energy']
df_opinion2011 = df_opinion2011.drop(['Code','Year'],axis=1)
```

```python
# pick 5 countries and worldwide
c = ['United States', 'China', 'Japan', 'Germany', 'France', 'World']
e = df_opinion2011['Country'].isin(c)
df_2011=df_opinion2011.loc[e]

ax=df_2011.set_index('Country').plot(kind='bar', stacked=True, color=['red', 'steelblue'], figsize=(10,6))

for bar in ax.patches:
    height = bar.get_height()
    width = bar.get_width()
    x = bar.get_x()
    y = bar.get_y()
    label_text = f'{height:.0f}%'
    label_x = x + width / 2
    label_y = y + height / 2

    if height>0:
      ax.text(label_x, label_y, label_text, ha='center',    
            va='center', color='white')


plt.gca().legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.title('2011 Public Opinion to Nuclear Energy Production', fontweight="bold")
plt.ylabel('% of Opinion')
plt.xticks(rotation=0)

plt.show()
```

Germany has the most people opposing to nuclear energy which is 79%. After the Fukushima Nuclear Plant Disaster, the public is more careful to safety issue, so the oppostation of most countries is more than supportion of nuclear energy.

# Indian Point Power Plant

## Figure 1:

**Source:** https://www.nuclearny.org/indian-point/

Illustrates that an 20% increase in fossil fuel consumption per terawatthour (TWh) occurred between May 2019 to November 2021 while nuclear energy consumption decreased 39% following the closure of two nuclear reactors at Indian Point Power Plant in New York. Indian Point's first nuclear reactor permanently shutdown in April 2020 and the second reactor permanently shutdown in April 2021, resulting in a lost of approximately 12 TWh of carbon-free nuclear energy. **Consequently, fossil fuels consumption (primarily natural gas) increased to replace the lost electricity that was generated from Indian Point Power Plant, resulting in carbon emissions increasing by 12 million metric tons.**

```python
IPP = pd.read_csv("NY_IPP_Change.csv",index_col=0)
fig, ax = plt.subplots(1, figsize=(10,10))

#plot
plt.plot(IPP["Renewables"], 'b', marker='o', markersize=5, linewidth=3)
plt.plot(IPP["Nuclear"], 'g', marker='o', markersize=5, linewidth=3)
plt.plot(IPP["Fossil Fuels"], 'r', marker='o', markersize=5, linewidth=3)
plt.title("NYS Electricity Mix from May 2019 to November 2021 (TWh)",  fontweight="bold")
plt.xlabel("Year", fontsize=12)
plt.xlim(2018.5,2021.5)
plt.xticks([2019, 2021], fontsize=12)
plt.yticks([])
ax.legend(IPP, bbox_to_anchor=(1.1,0.9), loc="upper right")

#data labels
x19=[21.7,27.2,32.4]
x21=[21.5,16.5,39.1]
percent=["+21% ", "-39% "]
location=[37,23]

for i in x19: #Start labels
    plt.annotate(i, #this is the text
    (2019,i), #these are the coordinates to position the label
    textcoords="offset points", #how to position the text
    xytext=(-15,-5), #distance from text to points (x,y)
    ha='center') #horizontal alignment can be left, right or center
    
for i in x21: #End labels
    plt.annotate(i, (2021,i), textcoords="offset points", xytext=(15,-5), ha='center')
    
for i, j in zip(percent, location): #Center Labels
    plt.annotate(i, (2020,j), textcoords="offset points", xytext=(2,-10), ha='center', fontweight="bold", fontsize=12)
    
#grid
ax.xaxis.grid(color='black', linestyle='solid', 
              which='both', alpha=0.9)

#remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
